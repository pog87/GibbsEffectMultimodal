```python
import pandas as pd
```

```python
df = pd.read_excel("pointwise_oversampling_DSSIM.xlsx", engine='openpyxl')
```

```python
df["Software"] = [a.split("_")[0] for a in df.name]
df["Image"]    = [a.split("_")[1] for a in df.name]
df["method"]   = [a.split("_")[-1] for a in df.name]
```

```python
df.columns
```

```python
#df.perc_volume *= 100
```

```python
df = df[['Image', 'Software', 'method',  'error', 'error_in_borders', 'percentage', 'perc_gradient','perc_volume' ]]
```

```python
df = df.replace('reg', 'ATrilinear')
df = df.replace('gauss', 'Gaussian')
df = df.replace('lancz', 'Lanczos')
df = df.replace('splines', 'Splines')
```

```python
df = df.sort_values(by = ['Image', 'Software', 'method'])
```
