```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.linspace(0,1, 101)

f = lambda x: np.exp(-2**2 * x**2)
```

```python
plt.plot(x,f(x))
```

```python
plt.plot(x,f(x) + f(1-x))
```

```python
f(.5)*2
```

```python
np.min(f(x) + f(1-x))
```

# Cubic Splines

```python
x = np.linspace(0,1, 101)
f1 = lambda x: x**3 - 2*x**2 +1
```

```python
plt.plot(x,f1(x))
```

```python
plt.plot(x,f1(x) + f1(1-x))
```

```python
f1(.5)*2, np.min(f1(x) + f1(1-x))
```

```python
x = np.linspace(0,1, 101)
f2 = lambda x: -x**3 +5*x**2 -8*x + 4
```

```python
plt.plot(x,f2(x+1))
```

```python
plt.plot(x,f2(1+x) + f2(2-x))
```

```python
np.min(f2(1+x) + f2(2-x)), 2*f2(1.5)
```

# Lanczos-2

```python
x = np.linspace(0,1, 101)
f1 = lambda x: np.sinc(x) * np.sinc(x/2)
```

```python
plt.plot(x,f1(x))
```

```python
plt.plot(x,f1(x) + f1(1-x))
```

```python
f1(.5)*2, np.min(f1(x) + f1(1-x))
```

```python
f2=f1
```

```python
plt.plot(x,f2(x+1))
```

```python
plt.plot(x,f2(1+x) + f2(2-x))
```

```python
np.min(f2(1+x) + f2(2-x)), 2*f2(1.5)
```

```python
1 + 2*f2(1.5)
```
